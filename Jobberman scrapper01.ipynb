```python
print("SPECIFY A KEYWORD")
keyword  = input(">")
print(f"Loading {keyword} jobs.......\n")
for page in range(0,7):
    data = requests.get("https://www.jobberman.com/jobs/software-data" +"?page=" +str(page))
    soup = BeautifulSoup(data.text,'html.parser')
    jobs = soup.find_all('div',class_="mx-5 md:mx-0 flex flex-wrap col-span-1 mb-5 bg-white rounded-lg border border-gray-300 hover:border-gray-400 focus-within:ring-2 focus-within:ring-offset-2 focus-within:ring-gray-500")
    for job in jobs:
        job_title = job.find('p',class_="text-lg font-medium break-words text-link-500").text
        if keyword in job_title:
            company_name = job.find('p',class_="text-sm text-link-500").text
            date = job.find('p',class_="ml-auto text-sm font-normal text-gray-700 text-loading-animate").text
            more_info = job.find('div',class_="flex flex-wrap mt-3 text-sm text-gray-500 md:py-0").text
            link = job.find('a',href = True)
            print(job_title.strip())
            print(company_name.strip())
            print(date.strip())
            print(more_info.strip())
            print(f"{link.get('href')}\n")
else:
    print(f"No {keyword} jobs......")
```
